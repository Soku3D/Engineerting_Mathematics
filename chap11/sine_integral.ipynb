```python
import scipy.special as special
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2,2,0.01)
a= 8
```

```python
f, arr = plt.subplots(3,figsize=(5,15))

for i in range(3):
  y1, y =  special.sici(a*(i+1)*(x+1))
  y3, y4 = special.sici(a*(i+1)*(x-1))
  final_val = (y1-y3)/np.pi
  arr[i].grid(True)
  arr[i].plot(x,final_val)
 


plt.show()
```
