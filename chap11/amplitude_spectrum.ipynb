**amplitude spectrum**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,400,0.01)
```

```python
y= (2*np.sin(x*np.pi))/(x*np.pi)
y2 =(2*np.sin(x*np.pi/4))/(x*np.pi)
y3 =(2*np.sin(x*np.pi/8))/(x*np.pi)
y4 =(2*np.sin(x*np.pi/16))/(x*np.pi)
y5 = (2*np.sin(x*np.pi/500))/(x*np.pi)
```

```python
plt.plot(x,y)
plt.grid(True)
plt.xlim([0,10]) 
plt.show()
```
